```python
df['total_bedrooms'].fillna(df['total_bedrooms'].mean(), inplace=True)
df.isnull().sum()
```

```python
df['households'].fillna(df['households'].median(), inplace=True)
df.isnull().sum()
```

```python
X = df['median_house_value'].fillna(df['median_house_value'].mean(), inplace=True)
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['ocean_proximity'] = encoder.fit_transform(df['ocean_proximity'])
df['ocean_proximity'].fillna(df['ocean_proximity'].mean(), inplace=True)
df.isnull().sum()
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

X = df.drop('ocean_proximity', axis=1)
y = df['ocean_proximity']

# Check the distribution of the target variable
print(y.value_counts())

# Apply SMOTE to the data
# Set k_neighbors to a value that fits your smallest class
smote = SMOTE(k_neighbors=2)  # Adjust k_neighbors as needed
X_resampled, y_resampled = smote.fit_resample(X, y)

# Check the distribution of the classes after SMOTE
print('Original dataset shape:', Counter(y))
print('Resampled dataset shape:', Counter(y_resampled))

```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df = scaler.fit_transform(df)
df = pd.DataFrame(df)
df
```

```python
df.drop_duplicates()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
x=df.drop('median_house_value',axis=1)
y=df['median_house_value']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
linear_reg = LinearRegression()
linear_reg.fit(x_train, y_train)
y_pred = linear_reg.predict(x_test)
r2 = r2_score(y_pred, y_test)
mse = mean_squared_error(y_pred, y_test)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df = scaler.fit_transform(df)
df = pd.DataFrame(df)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import math
x=df.drop(8,axis=1)
y=df[8]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
linear_reg = LinearRegression()
linear_reg.fit(x_train, y_train)
y_pred = linear_reg.predict(x_test)
r2 = r2_score(y_pred, y_test)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
print("Root mean squared error :", math.sqrt(mse))
```
